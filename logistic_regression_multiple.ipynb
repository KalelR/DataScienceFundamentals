# Quick and dirty logistic regression for multiple predictors
Example from ChatGPT. Just running to see that this works. I'll come back later and explore this further.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
```

```python
# Step 1: Load iris dataset
iris = load_iris()
X = iris.data
y = iris.target
print(X)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
print(X_normalized)
feats1 = [X[1] for X in X_normalized]
import numpy as np
print(f"mean = {np.mean(feats1)}, std = {np.std(feats1)}")
```

```python
# Step 2: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=0)

# Step 3: Instantiate the logistic regression model
logistic_regression_model = LogisticRegression()

# Step 4: Fit the model to the training data
logistic_regression_model.fit(X_train, y_train)

# Step 5: Make predictions on the test data
predictions = logistic_regression_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, predictions)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# Print classification report
print("Classification Report:")
print(classification_report(y_test, predictions, target_names=iris.target_names))
```

```python

# Evaluate the model
accuracy = logistic_regression_model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```
